```python
#либы и подключение
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('C:\\Users\\dimag\\Desktop\\Mash_learning\\RK\\heart.csv')
data.head()
```

```python
#Проверяем пропуски
data.isnull().sum()
```

Заметим, что пропусков нет, следовательно удалять ничего не нужно

```python
#Строим корреляционную матрицу
#Основным параметром для нас будет параметр "target" (1 = disease, 0 = no disease)
data.corr()
```

```python
#Визуализируем матрицу
sns.heatmap(data.corr())
```

cp: chest pain type
-- Value 0: asymptomatic
-- Value 1: atypical angina
-- Value 2: non-anginal pain
-- Value 3: typical angina
thalach: maximum heart rate achieved

```python
#Можно заметить, что наиболее сильная корреляция проявляется с параметрами cp и thalach
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['cp'],y=data['thalach'],hue=data['target'])
plt.xlabel('cp')
plt.ylabel('thalach')
plt.show()
```

```python
#На следующей диаграмме мы можем исследовать, сколько больных и здоровых людей было выявлено при определенном пульсе 
sns.factorplot('thalach', kind='count', hue='target', data=data, palette='coolwarm', height=30, aspect=.8)
```

```python
#Сделаем аналогичный график с симптомами
sns.factorplot('cp', kind='count', hue='target', data=data, palette='coolwarm', height=30, aspect=.8)
```

Можно заметить, что больше всего здоровых людей не имеют симптомов болезни (логично)
Построим график совместного распределения переменных cp и thalach для больных людей (доп требование)

```python
data1=data[data['target']==1]
sns.jointplot(x='cp', y='thalach', data=data1, kind='scatter');
```
